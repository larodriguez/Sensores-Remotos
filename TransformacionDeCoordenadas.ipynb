#### Conversions between different coordinate systems and datums

Coordinates on the Earth can be expressed using various datums, all related to each other through geometrical transformations.
Recall: A geodetic datum = size and shape of the Earth (= “coordinate system” + origin and orientation of the coordinate system).
A minimum of 5 parameters are necessary to define a datum:
- The translation of its origin with respect to the Earth’s center of mass (tX, tY, tZ)
- The size and shape of the Earth, usually approximated as an ellipsoid: semi-major axis a and flattening f (or, semi-major axis a and semi-minor axis b). There are hundreds of datums being used throughout the world, and it’s important to use coordinates that are associated with a well-defined (and known) datum. Using the wrong datum to interpret your coordinates can result in position errors of up to 1 km.
Simplest system: Earth Centered Earth Fixed cartesian coordinate system (ECEF). Origin = center of mass of the Earth, and coordinates are expressed as values for each of 3 right-handed orthogonal axes X, Y, Z (units are in meters). Z = rotation axis of the Earth, X-Y plane is the equatorial plane.

A more convenient (and intuitive) system is to express coordinates as latitude, longitude and height with respect to a given ellipsoid. The Prime Meridian is the origin for longitudes, and the Equator is the origin for latitudes.
- “Geodetic latitude” is defined as the angle from the equatorial plane to the vertical direction of the normal vector to the reference ellipsoid (This ellipsoid is the “datum”)
- “Geodetic longitude” of a point is the angle between a reference plane and a plane passing through the point, with both planes perpendicular to the equatorial plane.
- “Geodetic height” (also called “ellipsoid height”) at a point is the distance from the reference ellipsoid to the point in a direction normal to the surface of the ellipsoid.
